```python
zip_path = '/content/drive/MyDrive/Kampus/bahasaisyarat/dataset4.zip'
extract_path = 'dataset3'

if not os.path.exists(extract_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print(f"Dataset extracted to {extract_path}")
else:
    print(f"Dataset already extracted at {extract_path}")
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

model.train(
    data="/content/dataset3/data.yaml",
    epochs=30,
    imgsz=640,
    patience=15,
    batch=8
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("runs/detect/train/results.csv")

plt.figure(figsize=(12, 6))
plt.plot(df["epoch"], df["metrics/precision(B)"], label="Precision")
plt.plot(df["epoch"], df["metrics/recall(B)"], label="Recall")
plt.plot(df["epoch"], df["metrics/mAP50(B)"], label="mAP@0.5")
plt.plot(df["epoch"], df["metrics/mAP50-95(B)"], label="mAP@0.5:0.95")

plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Model Accuracy over Epochs")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["epoch"], df["train/box_loss"], label="Box Loss (Train)")
plt.plot(df["epoch"], df["train/cls_loss"], label="Class Loss (Train)")
plt.plot(df["epoch"], df["train/dfl_loss"], label="DFL Loss (Train)")
plt.plot(df["epoch"], df["val/box_loss"], label="Box Loss (Val)")
plt.plot(df["epoch"], df["val/cls_loss"], label="Class Loss (Val)")
plt.plot(df["epoch"], df["val/dfl_loss"], label="DFL Loss (Val)")

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training & Validation Loss over Epochs")
plt.legend()
plt.grid(True)
plt.show()
```
